```r
# BILL
bill <- function(quantity,price,gst){
   total_amount <-price*quantity
 gst_amount <- total_amount*gst/100
 total_price <- total_amount + gst_amount
 cat("\n................bill...........\n")
 cat("\nprice :",price,"/-")
 cat("\nquantity :",quantity, "/-")
 cat("\nGST :",gst,"%")
 cat("\nGST_amount =",gst_amount,"/-")
 cat("\ntotal_amount =",total_amount,"/-")
 cat("\ntotal_price(with GST) =",total_price,"/-")
}
```

```r
bill(4,450,12)
bill(12,8757,6)
bill(67,67576,8)
```
